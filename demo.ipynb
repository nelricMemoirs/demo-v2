
Permainan tongkat estafet diikuti sebanyak N peserta,
Setiap pemain membutuhkan waktu 1 detik untuk men-transfer tongkat ke pemain berikutnya, jika tongkat mencapai akhir pemain maka pemain terakhir harus memberikan tongkat ke pemain sebelumnya dengan urutan terbalik,
demi melakukan permainan yang adil/fair , wasit ingin melihat posisi tongkat pada t second , maka akan keluar siapa yang memegang tongkat pada waktu tersebut.

```
N = total pemain
t = waktu dalam second 
no. pemain = 1 s/d N-1

contoh input : N = 5, dan t = 4 

output : pemain = 1, 2, 3, 4, 5
         posisi tongkat pada detik ke 4 = pemain ke-3


contoh input : N = 5, dan t = 8 

output : pemain = 1, 2, 3, 4, 5
         posisi tongkat pada detik ke 8 = pemain ke- 1
```

````python


def pass_baton_2(n : int, time : int) -> int:
    
    player_sequence = list(range(1, n + 1)) + list(range(n - 1, 0, -1))
    
    # if time == len(player_sequence) or time > len(player_sequence) :
    #     return "game end"
    
    counter = 1
    for player in range(len(player_sequence)):
        # print(player_sequence[player])
        if time == counter :
            return player_sequence[player + 1]
        
        counter = counter + 1
    
    

    
````
